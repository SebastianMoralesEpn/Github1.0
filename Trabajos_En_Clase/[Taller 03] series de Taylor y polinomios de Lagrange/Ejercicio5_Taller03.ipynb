##### **Encuentre el polinomio de Lagrange para los siguientes datos y grafique:**

#### 5. $(1,1), (2,2), (3,2)$

```python
import numpy as np
import matplotlib.pyplot as plt

def polinomio(x):
    return (-x**2 + 5*x -2) / 2

points = [(1,1), (2,2), (3,2)]
x_lagrange = np.linspace(0.5, 4, 400)
y_lagrange = polinomio(x_lagrange)

plt.plot(x_lagrange, y_lagrange, label='Polinomio de Lagrange')
x_coords = [p[0] for p in points]
y_coords = [p[1] for p in points]
plt.plot(x_coords, y_coords, 'o', label='Puntos dados')
plt.legend()
plt.grid(True)
plt.title('Polinomio de Lagrange para (1,1), (2,2), (3,2)')
plt.show()
```
