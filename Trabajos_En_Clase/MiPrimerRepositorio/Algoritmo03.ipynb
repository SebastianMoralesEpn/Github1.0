# Algoritmo 03

```python
def algoritmo03(n):
    if n == 0:
        return 0
    else:
        x = 0
        y = 1
        for i in range(1, n):
            z = x + y
            x = y
            y = z
        return y
```

```python
# Caso de Prueba del Algoritmo con n = 11
n = 11
resultado = algoritmo03(n)
print(f"Para n = {n}, y = {resultado}")
```

```python
# Medici√≥n del tiempo de ejecuci√≥n para el Algoritmo03
import time

def algoritmo_03(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1

    x, y = 0, 1
    for _ in range(1, n):
        x, y = y, x + y
    return y

start_t = time.time()
[algoritmo_03(x) for x in range(5000)]
end_t = time.time()
print("Tiempo Algoritmo_03:", end_t - start_t)
```

```python
# Versi√≥n optimizada utilizando memoria previa
def fibonacci(limit):
    cache = {0: 0, 1: 1}
    x, y = 0, 1
    for i in range(1, limit):
        x, y = y, x + y
        cache[i + 1] = y  # Guardamos los valores precomputados
    return cache

cache = fibonacci(5000)  # Valores hasta 5000

def algoritmo_03_02(n):
    return cache[n]  # Recuperaci√≥n instant√°nea desde la cach√©

# Medici√≥n del tiempo de ejecuci√≥n
start_t = time.time()
[algoritmo_03_02(x) for x in range(5000)]
end_t = time.time()
print("Tiempo Algoritmo_03_02 (optimizado):", end_t - start_t)
```

```python
# Ejercicio Quizz (Determiar la iteraci√≥n m√≠nima donde el error relativo est√° dentro de 10^-5)

import math

def FL_rounding(num: float, digits: int) -> float:
    '''Redondea el n√∫mero ``num`` a ``digits`` cifras significativas.'''
    num = float(f"{num:.{digits - 1}e}")
    return num

def algoritmo03_quizz(n):
    fib = [0, 1]  
    for i in range(2, n + 1):
        fib.append(fib[i - 1] + fib[i - 2])
    return fib  # Retorna la serie completa hasta n

max_n = 5000  
fib_cache = algoritmo03_quizz(max_n)

# Valor de la raz√≥n dorada
phi = (1 + math.sqrt(5)) / 2

# Encontrar la iteraci√≥n m√≠nima con error relativo dentro de 10^-5
i = 2  # Empezamos desde el segundo t√©rmino
while i < max_n:
    ratio = FL_rounding(fib_cache[i + 1] / fib_cache[i], 3)  # Redondeo a 3 cifras
    error_relativo = abs((ratio - phi) / phi)

    if error_relativo < 10**-5:
        break  # Se encontr√≥ la iteraci√≥n correcta

    i += 1

print(f"La iteraci√≥n m√≠nima donde el error relativo est√° dentro de 10^-5 es: {i}")
```
