## Algoritmo 02

```python
def algortimo02(arr):
    n = len(arr)
    for i in range(n):
        swapped = False
        for j in range(1, n - i):
            if arr[j] < arr[j - 1]:
                arr[j], arr[j - 1] = arr[j - 1], arr[j]  # Intercambio de elementos
                swapped = True
        if not swapped:
            break  # Si no hubo intercambio, la lista ya est√° ordenada
    return arr

# Ejemplo de uso
arr = [64, 25, 12, 22, 11]
sorted_arr = algortimo02(arr)
print("Lista ordenada:", sorted_arr)
```

```python
# Ejercicio del Quizz (determinar el numero de comparaciones)
def algoritmo02(arr):
    n = len(arr)
    comparaciones = 0  # Se establece un contador de comparaciones
    for i in range(n):
        swapped = False
        for j in range(1, n - i):
            comparaciones += 1  # Incrementa contador por cada comparaci√≥n
            if arr[j] < arr[j - 1]:
                arr[j], arr[j - 1] = arr[j - 1], arr[j]  # Intercambio de elememtos
                swapped = True
        if not swapped:
            break  # Si no hubo intercambio, la lista ya est√° ordenada
    return arr, comparaciones

# Ejemplo Quizz
arr = [5, 4, 3, 2, 1]
sorted_arr, comparaciones = algoritmo02(arr)
print("Lista ordenada:", sorted_arr)
print("N√∫mero de comparaciones:", comparaciones)
```
