## **Splines cúbicos**

```python
def Spline(x: float, x0: float, pars: dict[str, float]) -> float:
    a = pars["a"]
    b = pars["b"]
    c = pars["c"]
    d = pars["d"]
    return a + b * (x - x0) + c * (x - x0) ** 2 + d * (x - x0) ** 3
```

```python
import matplotlib.pyplot as plt
import numpy as np

xs = [-1, 0, 1]
ys = [1, 5, 3]
s = [
    {"a": 1, "b": 5.5, "c": 0, "d": -1.5},
    {"a": 5, "b": 1, "c": -4.5, "d": 1.5},
]
for i, x_i in enumerate(xs[:-1]):
    _x = np.linspace(x_i, xs[i + 1], 20)
    _y = Spline(_x, x_i, s[i])
    plt.plot(_x, _y, color="red")

plt.scatter(xs, ys)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Interpolación con splines cúbicos")
plt.show()
```

Grafique los splines cúbicos tal que cumplan con una pendiente de m=-3 en los puntos dados.

```python
import numpy as np
from scipy.interpolate import CubicSpline
import matplotlib.pyplot as plt

# Datos de entrada
x = np.array([-1, 0, 1])
y = np.array([1, 5, 3])

# Condición de pendiente en x = -1: S0'(-1) = -3
m = -3

cs = CubicSpline(x, y, bc_type=((1, m), (2, 0)))

# Definir los splines S0 y S1
def S0(x_val):
    return cs(x_val)

def S1(x_val):
    return cs(x_val)

# Coeficientes de los splines (para mostrar en formato solicitado)
coeffs = cs.c

# Spline S0 (intervalo [-1, 0])
a0, b0, c0, d0 = coeffs[:, 0]
# Ajustar términos para la forma (x - xi)
xi_S0 = -1
S0_str = f"{y[0]} + {b0}(x - {xi_S0}) + {c0}(x - {xi_S0})^2 + {d0}(x - {xi_S0})^3"

# Spline S1 (intervalo [0, 1])
a1, b1, c1, d1 = coeffs[:, 1]
xi_S1 = 0
S1_str = f"{y[1]} + {b1}(x - {xi_S1}) + {c1}(x - {xi_S1})^2 + {d1}(x - {xi_S1})^3"

print("S0(x) =", S0_str)
print("S1(x) =", S1_str)

# Graficar los splines
x_vals = np.linspace(-1, 1, 100)
y_vals = cs(x_vals)

plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals, label='Splines cúbicos', color='blue')
plt.scatter(x, y, color='red', label='Puntos dados')
plt.title('Splines Cúbicos con Pendiente Especificada en x = -1')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
