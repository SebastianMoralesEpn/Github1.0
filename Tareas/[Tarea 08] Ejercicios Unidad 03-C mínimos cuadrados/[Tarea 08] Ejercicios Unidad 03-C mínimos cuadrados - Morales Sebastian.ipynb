**a. Construya el polinomio por mínimos cuadrados de grado 1 y calcule el error.**

```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

# Datos
xi = np.array([4.0, 4.2, 4.5, 4.7, 5.1, 5.5, 5.9, 6.3, 6.8, 7.1])
yi = np.array([102.56, 130.11, 113.18, 142.05, 167.53, 195.14, 224.87, 256.73, 299.50, 326.72])

# Modelo lineal
def linear_model(x, a, b):
    return a * x + b

# Ajuste
params_lin, _ = curve_fit(linear_model, xi, yi)
a1, b1 = params_lin
y_pred = linear_model(xi, a1, b1)
error = np.sum((yi - y_pred)**2)

# Gráfico
plt.figure(figsize=(10, 6))
plt.scatter(xi, yi, color='red', label='Datos originales')
x_vals = np.linspace(min(xi), max(xi), 100)
plt.plot(x_vals, linear_model(x_vals, a1, b1), 'b-', label=f'Ajuste lineal: y = {a1:.2f}x + {b1:.2f}')
plt.title('Ajuste lineal por mínimos cuadrados de Grado 1')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()

print("Polinomio de grado 1:")
print(f"y = {a1:.4f}x + {b1:.4f}")
print(f"Error cuadrático: {error:.4f}")
```

**b. Construya el polinomio de mínimos cuadrados de grado 2 y calcule el error.**

```python
# Modelo cuadrático
def quadratic_model(x, a, b, c):
    return a * x**2 + b * x + c

# Ajuste
params_quad, _ = curve_fit(quadratic_model, xi, yi)
a2, b2, c2 = params_quad
y_pred = quadratic_model(xi, a2, b2, c2)
error = np.sum((yi - y_pred)**2)

# Gráfico
plt.figure(figsize=(10, 6))
plt.scatter(xi, yi, color='red', label='Datos originales')
x_vals = np.linspace(min(xi), max(xi), 100)
plt.plot(x_vals, quadratic_model(x_vals, a2, b2, c2), 'g-', 
         label=f'Ajuste cuadrático: y = {a2:.2f}x² + {b2:.2f}x + {c2:.2f}')
plt.title('Ajuste cuadrático por mínimos cuadrados de Grado 2')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()

print("Polinomio de grado 2:")
print(f"y = {a2:.4f}x² + {b2:.4f}x + {c2:.4f}")
print(f"Error cuadrático: {error:.4f}")
```

**c. Construya el polinomio de mínimos cuadrados de grado 3 y calcule el error.**

```python
# Modelo cúbico
def cubic_model(x, a, b, c, d):
    return a * x**3 + b * x**2 + c * x + d

# Ajuste
params_cubic, _ = curve_fit(cubic_model, xi, yi)
a3, b3, c3, d3 = params_cubic
y_pred = cubic_model(xi, a3, b3, c3, d3)
error = np.sum((yi - y_pred)**2)

# Gráfico
plt.figure(figsize=(10, 6))
plt.scatter(xi, yi, color='red', label='Datos originales')
x_vals = np.linspace(min(xi), max(xi), 100)
plt.plot(x_vals, cubic_model(x_vals, a3, b3, c3, d3), 'm-',
         label=f'Ajuste cúbico: y = {a3:.2f}x³ + {b3:.2f}x² + {c3:.2f}x + {d3:.2f}')
plt.title('Ajuste cúbico por mínimos cuadrados de Grado 3')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()

print("Polinomio de grado 3:")
print(f"y = {a3:.4f}x³ + {b3:.4f}x² + {c3:.4f}x + {d3:.4f}")
print(f"Error cuadrático: {error:.4f}")
```

**d. Construya el polinomio de mínimos cuadrados de la forma be^ax y calcule el error.**

```python
# Modelo be^(ax)
def model_d(x, b, a):
    return b * np.exp(a * x)

# Ajuste
params_d, _ = curve_fit(model_d, xi, yi, p0=[1, 0.1])
bd, ad = params_d
y_pred = model_d(xi, bd, ad)
error = np.sum((yi - y_pred)**2)

# Gráfico
plt.figure(figsize=(10, 6))
plt.scatter(xi, yi, color='red', label='Datos originales')
plt.plot(x_vals, model_d(x_vals, bd, ad), 'c-', label=f'Ajuste: y = {bd:.4f}e^({ad:.4f}x)')
plt.title('Ajuste con modelo y = be^(ax)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()

print("Modelo be^(ax):")
print(f"y = {bd:.4f}e^({ad:.4f}x)")
print(f"Error cuadrático: {error:.4f}")
```

**3. La siguiente tabla muestra los promedios de puntos del colegio de 20 especialistas en matemáticas y ciencias computacionales, junto con las calificaciones que recibieron estos estudiantes en la parte de matemáticas de la prueba ACT (Programa de Pruebas de Colegios Americanos) mientras estaban en secundaria. Grafique estos datos y encuentre la ecuación de la recta por mínimos cuadrados para estos datos.**

```python
from scipy.stats import linregress

# Datos proporcionados
act_scores = np.array([28, 25, 28, 27, 28, 33, 28, 29, 23, 27, 29, 28, 27, 29, 21, 28, 28, 26, 30, 24])
gpa_scores = np.array([3.84, 3.21, 3.23, 3.63, 3.75, 3.20, 3.41, 3.38, 3.53, 2.03, 3.75, 3.65, 3.87, 3.75, 1.66, 3.12, 2.96, 2.92, 3.10, 2.81])

# Realizamos la regresión lineal
slope, intercept, r_value, p_value, std_err = linregress(act_scores, gpa_scores)

# Calculamos los valores predichos
predicted_gpa = intercept + slope * act_scores

# Gráfico de dispersión y línea de regresión
plt.figure(figsize=(12, 7))
plt.scatter(act_scores, gpa_scores, color='blue', label='Datos reales', alpha=0.7)
plt.plot(act_scores, predicted_gpa, color='red', label=f'Regresión lineal: y = {slope:.3f}x + {intercept:.3f}')

# Personalización del gráfico
plt.title('Relación entre Puntuación ACT y Promedio de Puntos', fontsize=14)
plt.xlabel('Puntuación ACT', fontsize=12)
plt.ylabel('Promedio de Puntos (GPA)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(fontsize=12)
plt.xticks(np.arange(20, 35, 1))
plt.yticks(np.arange(1.5, 4.1, 0.5))
plt.show()

# Resultados estadísticos
print(f"Ecuación: y = {slope:.4f}x + {intercept:.4f}")
print(f"Error: {std_err:.4f}")
```

**4. El siguiente conjunto de datos, presentando el Subcomité Antimonopolio del Senado, muestra las características comparativas de supervivencia durante un choque de automóviles de diferentes clases. Encuentre la recta por mínimos cuadrados que aproxima estos datos (la tabla muestra el porcentaje de vehículos que participaron en un accidente en los que la lesión más grave fue fatal o seria)**

```python
# Datos proporcionados
pesos = np.array([4800, 3700, 3400, 2800, 1900])  # en libras
porcentajes = np.array([3.1, 4.0, 5.2, 6.4, 9.6])  # porcentaje de lesiones graves/fatales

# Realizamos la regresión lineal
slope, intercept, r_value, p_value, std_err = linregress(pesos, porcentajes)

# Calculamos los valores predichos
predicted_percent = intercept + slope * pesos

# Gráfico de dispersión y línea de regresión
plt.figure(figsize=(12, 7))
plt.scatter(pesos, porcentajes, color='red', s=100, label='Datos reales')
plt.plot(pesos, predicted_percent, color='blue', label=f'Regresión lineal: y = {slope:.5f}x + {intercept:.2f}')

# Personalización del gráfico
plt.title('Relación entre Peso del Vehículo y Porcentaje de Lesiones Graves/Fatales', fontsize=14)
plt.xlabel('Peso del vehículo (lb)', fontsize=12)
plt.ylabel('Porcentaje de lesiones graves/fatales', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(fontsize=12, loc='upper right')
plt.show()

# Resultados estadísticos
print(f"Ecuación de la recta: y = {slope:.6f}x + {intercept:.2f}")
print(f"Error estándar: {std_err:.4f}")
```
