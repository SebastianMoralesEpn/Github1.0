```python
import numpy as np

# Definimos la función
def equation(x):
    return x**3 - 3*x**2 + x - 1

# Método de la bisección
def bisection_method(func, a, b, tol=1e-6, max_iter=100):
    if func(a) * func(b) >= 0:
        raise ValueError("La función debe cambiar de signo en el intervalo [a, b].")
    
    iter_count = 0
    while (b - a) / 2 > tol and iter_count < max_iter:
        c = (a + b) / 2
        if func(c) == 0:  # Encontramos la raíz exacta
            return c
        elif func(a) * func(c) < 0:
            b = c
        else:
            a = c
        iter_count += 1
    
    return (a + b) / 2  # Retornamos la aproximación de la raíz

# Intervalos iniciales donde se sospecha que hay raíces
intervals = [(-2, 0), (0, 2), (2, 4)]

# Encontrar raíces usando el método de la bisección
roots = []
for a, b in intervals:
    try:
        root = bisection_method(equation, a, b)
        roots.append(root)
    except ValueError as e:
        print(f"Error en el intervalo [{a}, {b}]: {e}")

print("Raíces encontradas:", roots)
```

```python
from scipy.optimize import newton

# Definimos la función
def f(x):
    return x**3 - 3*x**2 + x - 1

# Definimos la derivada de la función
def f_prime(x):
    return 3*x**2 - 6*x + 1

# Usamos la función de Newton con argumentos adicionales y salida completa
root_newton_args = newton(f, x0=100, fprime=f_prime, args=(), full_output=True)


# Usamos la función de Newton y contamos las iteraciones
iteraciones_newton_args = []

def f_count_args(x):
    iteraciones_newton_args.append(x)
    return f(x)

# Ejecutamos el método de Newton con la función que cuenta iteraciones
root_newton_args = newton(f_count_args, x0=100, fprime=f_prime, args=(), full_output=True)

# Mostramos los resultados de cada iteración
for i, val in enumerate(iteraciones_newton_args, 1):
    print(f"Iteración {i}: x = {val}")

# Mostramos los resultados finales
print("Raíz encontrada:", root_newton_args[0])
print("Convergencia:", root_newton_args[1].converged)
print("Iteraciones realizadas:", root_newton_args[1].iterations)
```
