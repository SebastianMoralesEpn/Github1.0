**Literal a.**

```python
import numpy as np
import matplotlib.pyplot as plt

# Sistema a)
x = np.linspace(-5, 5, 100)
y1 = (-x)/2  # x₁ + 2x₂ = 0 → x₂ = -x₁/2
y2 = x       # x₁ - x₂ = 0 → x₂ = x₁

plt.figure(figsize=(8, 6))
plt.plot(x, y1, label='x₁ + 2x₂ = 0')
plt.plot(x, y2, label='x₁ - x₂ = 0')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid()
plt.legend()
plt.title('Sistema a)')
plt.xlabel('x₁')
plt.ylabel('x₂')
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.show()
```

```python
# Sistema b)
y1 = (3 - x)/2      # x₁ + 2x₂ = 3 → x₂ = (3 - x₁)/2
y2 = (-6 - 2*x)/4   # -2x₁ - 4x₂ = 6 → x₂ = (-6 - 2x₁)/4

plt.figure(figsize=(8, 6))
plt.plot(x, y1, label='x₁ + 2x₂ = 3')
plt.plot(x, y2, label='-2x₁ - 4x₂ = 6')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid()
plt.legend()
plt.title('Sistema b)')
plt.xlabel('x₁')
plt.ylabel('x₂')
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.show()
```

```python
# Definimos el rango de x₁
x = np.linspace(-5, 5, 100)

# Ecuación 1: 2x₁ + x₂ = -1 → x₂ = -1 - 2x₁
y1 = -1 - 2*x

# Ecuación 2: x₁ + x₂ = 2 → x₂ = 2 - x₁
y2 = 2 - x

# Ecuación 3: x₁ - 3x₂ = 5 → x₂ = (x₁ - 5)/3
y3 = (x - 5)/3

# Configuración del gráfico
plt.figure(figsize=(10, 8))
plt.plot(x, y1, label='2x₁ + x₂ = -1', color='blue')
plt.plot(x, y2, label='x₁ + x₂ = 2', color='green')
plt.plot(x, y3, label='x₁ - 3x₂ = 5', color='red')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True, linestyle='--', alpha=0.7)

# Encontrar puntos de intersección
# Intersección entre ecuación 1 y 2
A = np.array([[2, 1], [1, 1]])
b1 = np.array([-1, 2])
sol1 = np.linalg.solve(A, b1)

# Intersección entre ecuación 1 y 3
A = np.array([[2, 1], [1, -3]])
b2 = np.array([-1, 5])
sol2 = np.linalg.solve(A, b2)

# Intersección entre ecuación 2 y 3
A = np.array([[1, 1], [1, -3]])
b3 = np.array([2, 5])
sol3 = np.linalg.solve(A, b3)

# Marcar puntos de intersección
plt.scatter(sol1[0], sol1[1], color='purple', zorder=5, label=f'Intersección 1-2 ({sol1[0]:.1f}, {sol1[1]:.1f})')
plt.scatter(sol2[0], sol2[1], color='orange', zorder=5, label=f'Intersección 1-3 ({sol2[0]:.1f}, {sol2[1]:.1f})')
plt.scatter(sol3[0], sol3[1], color='brown', zorder=5, label=f'Intersección 2-3 ({sol3[0]:.1f}, {sol3[1]:.1f})')

# Configuración adicional del gráfico
plt.legend()
plt.title('Sistema c) - Tres ecuaciones con dos incógnitas')
plt.xlabel('x₁')
plt.ylabel('x₂')
plt.xlim(-3, 5)
plt.ylim(-4, 6)
plt.show()

# Verificar si hay solución común a las tres ecuaciones
try:
    A = np.array([[2, 1], [1, 1], [1, -3]])
    b = np.array([-1, 2, 5])
    # Usamos mínimos cuadrados para sistemas sobredeterminados
    sol, residuals, _, _ = np.linalg.lstsq(A, b, rcond=None)
    print(f"\nSolución por mínimos cuadrados: x₁ = {sol[0]:.2f}, x₂ = {sol[1]:.2f}")
    print(f"Residuos (errores): {residuals}")
except np.linalg.LinAlgError:
    print("\nNo existe una solución exacta para las tres ecuaciones simultáneamente")
```
