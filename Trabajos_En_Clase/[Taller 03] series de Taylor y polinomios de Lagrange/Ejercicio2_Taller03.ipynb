**Grafique las curvas de las series de Taylor de varios órdenes para los siguientes casos:**

```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos el rango de x (evitando x=1 donde la función tiene una asíntota)
x = np.linspace(-1.5, 0.9, 100)

# Definimos la función exacta
y_exact = 1 / (1 - x)

# Definimos las aproximaciones de Taylor
taylor_0 = np.ones_like(x)  # Orden 0: 1
taylor_1 = 1 + x             # Orden 1: 1 + x
taylor_2 = 1 + x + x**2      # Orden 2: 1 + x + x²
taylor_3 = 1 + x + x**2 + x**3  # Orden 3: 1 + x + x² + x³

# Configuramos la gráfica
plt.figure(figsize=(10, 6))
plt.plot(x, y_exact, label='1/(1-x) exacta', color='blue', linewidth=2)
plt.plot(x, taylor_0, label='Orden 0: 1', linestyle='--', color='red')
plt.plot(x, taylor_1, label='Orden 1: 1 + x', linestyle='--', color='green')
plt.plot(x, taylor_2, label='Orden 2: 1 + x + x²', linestyle='--', color='orange')
plt.plot(x, taylor_3, label='Orden 3: 1 + x + x² + x³', linestyle='--', color='purple')

# Añadimos detalles a la gráfica
plt.title('Aproximación de Taylor para 1/(1-x) en x₀=0')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)

# Establecemos límites apropiados
plt.xlim(-1.5, 0.9)
plt.ylim(-1, 10)  # Ajustamos para ver mejor las aproximaciones

# Mostramos la gráfica
plt.show()
```
