**1. Use el método de bisección para encontrar soluciones precisas dentro de 10^-2 para**

**en cada intervalo.**

**a.[0,1]**

```python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return x**3 - 7*x**2 + 14*x - 6

def biseccion(f, a, b, tol=1e-2, max_iter=1000):
    fa = f(a)
    fb = f(b)
    if fa*fb > 0:
        raise ValueError("La función no cambia de signo en el intervalo dado.")
    iter_count = 0
    while (b - a)/2 > tol and iter_count < max_iter:
        c = (a + b)/2
        fc = f(c)
        iter_count += 1
        if fc == 0:
            return c, iter_count
        if fa*fc < 0:
            b = c
            fb = fc
        else:
            a = c
            fa = fc
    return (a + b)/2, iter_count

# Intervalo dado
a, b = 0, 1
tolerancia = 1e-2
raiz, iteraciones = biseccion(f, a, b, tol=tolerancia)

print(f"Raíz encontrada: {raiz:.4f}")
print(f"Iteración en que se encontró la raíz: {iteraciones}")

# Graficar la función en el intervalo [0, 1]
x_vals = np.linspace(a, b, 400)
y_vals = f(x_vals)

plt.figure(figsize=(8,5))
plt.plot(x_vals, y_vals, label=r'$f(x) = x^3 - 7x^2 + 14x - 6$')
plt.axhline(0, color='black', linewidth=0.8)
plt.scatter(raiz, f(raiz), color='red', zorder=5, label=f'Raíz approx: {raiz:.4f}')
plt.title(f'Método de Bisección - Raíz en [{a}, {b}]')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```

**b. Use el método de bisección para encontrar una aproximación dentro de 10^-5 para un valor en [−2, 0] con 𝑥^2 − 1 = ℯ^(1-x^2)**

```python
# Definir la función para x^2 - 1 = e^(1 - x^2)
def h(x):
    return x**2 - 1 - np.exp(1 - x**2)

# Intervalo dado
a, b = -2, 0
tolerancia = 1e-5

# Encontrar la raíz usando el método de bisección
raiz, iteraciones = biseccion(h, a, b, tol=tolerancia)

print(f"Raíz encontrada: {raiz:.6f}")
print(f"Iteración en que se encontró la raíz: {iteraciones}")
print(f"Solución aproximada en el intervalo [{a}, {b}]: x ≈ {raiz:.6f}")
```
