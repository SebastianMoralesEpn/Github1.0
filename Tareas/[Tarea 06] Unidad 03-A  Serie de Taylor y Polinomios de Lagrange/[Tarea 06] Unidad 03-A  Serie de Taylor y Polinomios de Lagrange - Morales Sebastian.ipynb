**Determine el orden de la mejor aproximación para las siguientes funciones, usando la Serie de Taylor y el Polinomio de Lagrange:**

### **Serie de Taylor:**

1. 

```python
import numpy as np
import matplotlib.pyplot as plt

# Definición de la función objetivo
def funcion_meta(x):
    return 1 / (25 * x**2 + 1)

# Aproximación usando un polinomio de Taylor de orden 4
def taylor_aprox(x):
    return 1 - 25 * x**2 + 625 * x**4

# Valores del dominio
valores_x = np.linspace(-0.2, 0.2, 400)

# Evaluación de funciones
original = funcion_meta(valores_x)
aproximacion = taylor_aprox(valores_x)

# Gráfica comparativa
plt.figure(figsize=(8, 6))
plt.plot(valores_x, original, color='darkorange', linewidth=2, label='Función Base')
plt.plot(valores_x, aproximacion, color='royalblue', linestyle='--', linewidth=2, label='Taylor orden 4')
plt.title('Comparación de Función y Aproximación de Taylor')
plt.xlabel('Eje X')
plt.ylabel('Valor de la función')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

2. 

```python
import sympy as sp

# Definir variable simbólica y función objetivo
variable = sp.Symbol('x')
funcion_base = sp.atan(variable)
punto_expansion = 1  # x0

# Orden del polinomio de Taylor
orden = 4
serie_taylor = sp.series(funcion_base, variable, punto_expansion, orden + 1).removeO()

# Conversión a funciones numéricas
f_original = sp.lambdify(variable, funcion_base, 'numpy')
f_taylor = sp.lambdify(variable, serie_taylor, 'numpy')

# Rango de evaluación para la gráfica
dominio = np.linspace(-2, 4, 500)
valores_original = f_original(dominio)
valores_aproximados = f_taylor(dominio)

# Visualización de los resultados
plt.figure(figsize=(10, 6))
plt.plot(dominio, valores_original, label="Función arctan(x)", color="green", linewidth=2)
plt.plot(dominio, valores_aproximados, label=f"Taylor orden {orden} en x={punto_expansion}", color="crimson", linestyle="--", linewidth=2)
plt.axvline(x=punto_expansion, color="purple", linestyle=":", label=f"Expansión en x = {punto_expansion}")
plt.title("Comparación: Función vs Aproximación de Taylor")
plt.xlabel("Eje X")
plt.ylabel("Valor de la función")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

### **Polinomio de Lagrange:**

1. 

```python
# Función a interpolar
def funcion_objetivo(x):
    return 1 / (25 * x**2 + 1)

# Interpolación manual tipo Lagrange
def interpolacion_lagrange(x):
    return -5 * x**2 + 1  # Aproximación construida manualmente

# Puntos base para la interpolación
puntos_x = [0, -0.4, 0.4]
puntos_y = [1, 0.2, 0.2]

# Crear el rango de valores para graficar
x_grafico = np.linspace(-0.4, 0.4, 500)
y_funcion = funcion_objetivo(x_grafico)
y_interp = interpolacion_lagrange(x_grafico)

# Crear la gráfica
plt.figure(figsize=(10, 6))
plt.plot(x_grafico, y_funcion, color='darkgreen', linewidth=2, label="Función objetivo")
plt.plot(x_grafico, y_interp, color='orange', linestyle='--', linewidth=2, label="Interpolación Lagrange aproximada")
plt.scatter(puntos_x, puntos_y, color='purple', s=60, label="Puntos utilizados", zorder=5)
plt.title("Interpolación tipo Lagrange realizada manualmente")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

2. 

```python
# Definir función base
def funcion_arctan(x):
    return np.arctan(x)

# Aproximación de Taylor (orden 2) en x = 1
def aproximacion_taylor_2(x):
    valor_central = np.pi / 4  # arctan(1)
    return valor_central + 0.5 * (x - 1) - 0.25 * (x - 1)**2

# Rango de valores a graficar
x_dominio = np.linspace(0, 2, 400)
y_funcion = funcion_arctan(x_dominio)
y_taylor = aproximacion_taylor_2(x_dominio)

# Crear la gráfica
plt.figure(figsize=(9, 6))
plt.plot(x_dominio, y_funcion, label="Función arctan(x)", color="darkorange", linewidth=2)
plt.plot(x_dominio, y_taylor, label="Polinomio de Taylor (orden 2)", color="teal", linestyle='--', linewidth=2)
plt.scatter(1, funcion_arctan(1), color="magenta", s=60, zorder=5, label="Punto de expansión (x = 1)")
plt.title("Comparación entre arctan(x) y su Aproximación de Taylor (Orden 2)")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
