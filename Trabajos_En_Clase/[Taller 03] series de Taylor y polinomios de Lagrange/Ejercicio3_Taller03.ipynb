##### **Grafique las curvas de las series de Taylor de varios órdenes para los siguientes casos:**

##### 3. $ln(x)$, $x_0 = 1$

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0.15, 20, 150)

ln_aprox = (x - 1)
ln_aprox1 = (x - 1) - (1/2) * (x - 1)**2
ln_aprox2 = (x - 1) - (1/2) * (x - 1)**2 + (1/3) * (x - 1)**3
ln_aprox3 = (x - 1) - (1/2) * (x - 1)**2 + (1/3) * (x - 1)**3 - (1/4) * (x - 1)**4

plt.figure(figsize=(10, 6))
plt.plot(x, np.log(x), label='ln(x)', color='black')
plt.plot(x, ln_aprox, label='ln(x) = (x - 1)', color='blue', linestyle='--')
plt.plot(x, ln_aprox1, label= 'ln (x) = (x - 1) - (1/2) * (x-1)²', color='green', linestyle='--')
plt.plot(x, ln_aprox2, label='ln(x) = (x - 1) - (1/2) * (x-1)² + (1/3) * (x-1)³', color='orange', linestyle='--')
plt.plot(x, ln_aprox3, label='ln(x) = (x - 1) - (1/2) * (x-1)² + (1/3) * (x-1)³ - (1/4) * (x-1)⁴', color='red', linestyle='--')

plt.title('Taylor Series Approximation of ln(x)')
plt.xlabel('x')
plt.ylabel('ln(x)')
plt.legend()
plt.grid(True)
plt.xlim(-1, 6)
plt.ylim(-10, 10)
plt.show()
```
