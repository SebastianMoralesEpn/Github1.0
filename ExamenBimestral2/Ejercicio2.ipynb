```python
from scipy.optimize import bisect
import matplotlib.pyplot as plt
import numpy as np

def mi_funcion(x: float):
    # Modificada según la ecuación dada
    return x**5 + 5.5*x**4 + 4.5*x**3 - 9.5*x**2 - 15.5*x - 6

# Generamos valores de x para visualizar la función
x = np.linspace(-5, 5, 100)

# Mostramos la gráfica de la función
plt.plot(x, [mi_funcion(valor) for valor in x])
plt.axhline(0, color='black', linewidth=1)  # Línea en y=0
plt.axvline(0, color='black', linewidth=1)  # Línea en x=0
plt.title("Visualización de la función")
plt.xlabel("x")
plt.ylabel("mi_funcion(x)")
plt.grid(True)
plt.show()

# Verificamos los signos en los extremos de los intervalos
print("Signos en los extremos de los intervalos:")
print(f"mi_funcion(-5): {mi_funcion(-5)}, mi_funcion(-2): {mi_funcion(-2)}")
print(f"mi_funcion(-2): {mi_funcion(-2)}, mi_funcion(0): {mi_funcion(0)}")
print(f"mi_funcion(0): {mi_funcion(0)}, mi_funcion(2): {mi_funcion(2)}")
print(f"mi_funcion(2): {mi_funcion(2)}, mi_funcion(5): {mi_funcion(5)}")

# Buscamos raíces en los intervalos donde los signos son opuestos
try:
    raiz1 = bisect(mi_funcion, a=-5, b=-2)
    print(f"Raíz en el primer intervalo: {raiz1}")
except ValueError as e:
    print(f"No se encontró raíz en el primer intervalo: {e}")

try:
    raiz2 = bisect(mi_funcion, a=-2, b=0)
    print(f"Raíz en el segundo intervalo: {raiz2}")
except ValueError as e:
    print(f"No se encontró raíz en el segundo intervalo: {e}")

try:
    raiz3 = bisect(mi_funcion, a=0, b=2)
    print(f"Raíz en el tercer intervalo: {raiz3}")
except ValueError as e:
    print(f"No se encontró raíz en el tercer intervalo: {e}")

try:
    raiz4 = bisect(mi_funcion, a=2, b=5)
    print(f"Raíz en el cuarto intervalo: {raiz4}")
except ValueError as e:
    print(f"No se encontró raíz en el cuarto intervalo: {e}")
```
