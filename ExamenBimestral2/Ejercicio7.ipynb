## Comparación con la solución real
La solución real de la ecuación diferencial es:
$y(t)=t^2 +1/3e^{-5t}$

* Grafique la solución real y la aproximación obtenida con el método de Euler en una misma gráfica.



```python
import matplotlib.pyplot as plt

# Definir la función que representa la ecuación diferencial
def f(t, y):
    return -5*y + 5*t**2 + 2*t

# Establecer los parámetros iniciales
a = 0      # inicio del intervalo
b = 1      # fin del intervalo
y_t0 = 1/3 # valor inicial de y
N = 10     # cantidad de puntos

# Aplicar el método de Euler para aproximar la solución
ys, ts, h = ODE_euler(a=a, b=b, f=f, y_t0=y_t0, N=N)

# Mostrar el tamaño del paso
print(f"El valor de es h: {h}")

# Visualizar los resultados en una gráfica
plt.plot(ts, ys, label="Euler aproximado")
plt.xlabel("t")
plt.ylabel("y")
plt.title("Aproximación de la EDO con Euler")
plt.legend()
plt.grid(True)
plt.show()
```

# Errores
Calcule el eeror relativo para diferentes valores de N.

```python
# Calculo del error relativo

# Función que define la ecuación diferencial
def f(t, y):
    return -5*y + 5*t**2 + 2*t

# Solución analítica de la ecuación diferencial
def solucion_real(t):
    return t**2 + (1/3) * np.exp(-5*t)

# Función para obtener el error relativo promedio
def error_relativo_promedio(ys_aprox, ys_real):
    return np.mean([abs(y_aprox - y_real) / abs(y_real) for y_aprox, y_real in zip(ys_aprox, ys_real)])

# Parámetros iniciales
a = 0
b = 1
y_t0 = 1/3

# Lista de valores de N a analizar
valores_N = [10, 5, 20]

# Ciclo para resolver y comparar para cada N
for N in valores_N:
    ys, ts, h = ODE_euler(a=a, b=b, f=f, y_t0=y_t0, N=N)
    ys_real = [solucion_real(t) for t in ts]
    error_rel = error_relativo_promedio(ys, ys_real)
    print(f"N = {N}:")
    print(f"  Paso h: {h}")
    print(f"  Error relativo promedio: {error_rel}")
    plt.plot(ts, ys, label=f"Euler (N={N})", marker='o')
    plt.plot(ts, ys_real, label="Solución exacta", linestyle='--')

plt.xlabel("t")
plt.ylabel("y")
plt.title("Solución exacta vs. aproximación de Euler")
plt.legend()
plt.grid(True)
plt.show()
```
