**a. Construya el polinomio por mínimos cuadrados de grado 1 y calcule el error.**

```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

# Datos
xi = np.array([4.0, 4.2, 4.5, 4.7, 5.1, 5.5, 5.9, 6.3, 6.8, 7.1])
yi = np.array([102.56, 130.11, 113.18, 142.05, 167.53, 195.14, 224.87, 256.73, 299.50, 326.72])

# Modelo lineal
def linear_model(x, a, b):
    return a * x + b

# Ajuste
params_lin, _ = curve_fit(linear_model, xi, yi)
a1, b1 = params_lin
y_pred = linear_model(xi, a1, b1)
error = np.sum((yi - y_pred)**2)

# Gráfico
plt.figure(figsize=(10, 6))
plt.scatter(xi, yi, color='red', label='Datos originales')
x_vals = np.linspace(min(xi), max(xi), 100)
plt.plot(x_vals, linear_model(x_vals, a1, b1), 'b-', label=f'Ajuste lineal: y = {a1:.2f}x + {b1:.2f}')
plt.title('Ajuste lineal por mínimos cuadrados de Grado 1')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()

print("Polinomio de grado 1:")
print(f"y = {a1:.4f}x + {b1:.4f}")
print(f"Error cuadrático: {error:.4f}")
```

**b. Construya el polinomio de mínimos cuadrados de grado 2 y calcule el error.**

```python
# Modelo cuadrático
def quadratic_model(x, a, b, c):
    return a * x**2 + b * x + c

# Ajuste
params_quad, _ = curve_fit(quadratic_model, xi, yi)
a2, b2, c2 = params_quad
y_pred = quadratic_model(xi, a2, b2, c2)
error = np.sum((yi - y_pred)**2)

# Gráfico
plt.figure(figsize=(10, 6))
plt.scatter(xi, yi, color='red', label='Datos originales')
x_vals = np.linspace(min(xi), max(xi), 100)
plt.plot(x_vals, quadratic_model(x_vals, a2, b2, c2), 'g-', 
         label=f'Ajuste cuadrático: y = {a2:.2f}x² + {b2:.2f}x + {c2:.2f}')
plt.title('Ajuste cuadrático por mínimos cuadrados de Grado 2')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()

print("Polinomio de grado 2:")
print(f"y = {a2:.4f}x² + {b2:.4f}x + {c2:.4f}")
print(f"Error cuadrático: {error:.4f}")
```

**c. Construya el polinomio de mínimos cuadrados de grado 3 y calcule el error.**

```python
# Modelo cúbico
def cubic_model(x, a, b, c, d):
    return a * x**3 + b * x**2 + c * x + d

# Ajuste
params_cubic, _ = curve_fit(cubic_model, xi, yi)
a3, b3, c3, d3 = params_cubic
y_pred = cubic_model(xi, a3, b3, c3, d3)
error = np.sum((yi - y_pred)**2)

# Gráfico
plt.figure(figsize=(10, 6))
plt.scatter(xi, yi, color='red', label='Datos originales')
x_vals = np.linspace(min(xi), max(xi), 100)
plt.plot(x_vals, cubic_model(x_vals, a3, b3, c3, d3), 'm-',
         label=f'Ajuste cúbico: y = {a3:.2f}x³ + {b3:.2f}x² + {c3:.2f}x + {d3:.2f}')
plt.title('Ajuste cúbico por mínimos cuadrados de Grado 3')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()

print("Polinomio de grado 3:")
print(f"y = {a3:.4f}x³ + {b3:.4f}x² + {c3:.4f}x + {d3:.4f}")
print(f"Error cuadrático: {error:.4f}")
```

**d. Construya el polinomio de mínimos cuadrados de la forma be^ax y calcule el error.**

```python
# Modelo be^(ax)
def model_d(x, b, a):
    return b * np.exp(a * x)

# Ajuste
params_d, _ = curve_fit(model_d, xi, yi, p0=[1, 0.1])
bd, ad = params_d
y_pred = model_d(xi, bd, ad)
error = np.sum((yi - y_pred)**2)

# Gráfico
plt.figure(figsize=(10, 6))
plt.scatter(xi, yi, color='red', label='Datos originales')
plt.plot(x_vals, model_d(x_vals, bd, ad), 'c-', label=f'Ajuste: y = {bd:.4f}e^({ad:.4f}x)')
plt.title('Ajuste con modelo y = be^(ax)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()

print("Modelo be^(ax):")
print(f"y = {bd:.4f}e^({ad:.4f}x)")
print(f"Error cuadrático: {error:.4f}")
```

**e. Construya el polinomio de mínimos cuadrados de la forma bx^a y calcule el error.**

```python
# Modelo bx^a
def model_e(x, b, a):
    return b * x**a

# Ajuste
params_e, _ = curve_fit(model_e, xi, yi, p0=[1, 1])
be, ae = params_e
y_pred = model_e(xi, be, ae)
error = np.sum((yi - y_pred)**2)

# Gráfico
plt.figure(figsize=(10, 6))
plt.scatter(xi, yi, color='red', label='Datos originales')
x_vals = np.linspace(min(xi), max(xi), 100)
plt.plot(x_vals, model_e(x_vals, be, ae), 'y-',
         label=f'Ajuste: y = {be:.4f}x^{ae:.4f}')
plt.title('Ajuste con modelo y = bx^a')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()

print("Modelo bx^a:")
print(f"y = {be:.4f}x^{ae:.4f}")
print(f"Error cuadrático: {error:.4f}")
```
