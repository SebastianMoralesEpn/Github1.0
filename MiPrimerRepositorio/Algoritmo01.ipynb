## Algoritmo 01

```python
def algoritmo01(N, X):
    SUM = 0  
    for i in range(N):  
        SUM += X[i] 
    return SUM  

# Casos de prueba
X1 = list(range(100))  # Primer caso de prueba
resultado1 = algoritmo01(len(X1), X1)
print("Caso 1:", resultado1)

X2 = [1, 1/2, 1/4]  # Segundo caso de prueba
resultado2 = algoritmo01(len(X2), X2)
print("Caso 2:", resultado2)

# Generaci√≥n de la lista X
X = [1 / 2**x for x in range(15)]
print("Lista X:", X)

# Aplicaci√≥n del algoritmo
r = algoritmo01(len(X), X)
print("Resultado r:", r)

# C√°lculo de la diferencia con 2
diferencia = 2 - r
print(f"Diferencia 2 - r: {diferencia}")
```

```python
def algoritmo01(N, X):
    SUM = 0  
    for i in range(N):  
        SUM += X[i]  
        if 2 - SUM < 10**-1:  # Condici√≥n de error absoluto
            return SUM, i + 1  # Retorna la suma y el n√∫mero de elementos usados
    return SUM, N  

# Generaci√≥n de la lista X 
X = [1 / 2**x for x in range(100)]  # Se genera un n√∫mero grande de t√©rminos por precauci√≥n

# Aplicaci√≥n del algoritmo
r, min_elementos = algoritmo01(len(X), X)

print("Resultado r:", r)
print("N√∫mero m√≠nimo de elementos:", min_elementos)
print(f"Diferencia 2 - r: {2 - r}")
```
