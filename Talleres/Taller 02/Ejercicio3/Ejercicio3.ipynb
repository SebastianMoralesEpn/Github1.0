```python
import numpy as np

# Definir la función 
def funcion(x):
    return np.sin(x) / x if x != 0 else 1.0

# Establecer el umbral mínimo permitido
limite = 0.015

# Precisión de la búsqueda
incremento = 0.01
punto = 1.0  # Empezar desde 1 para evitar división por cero

# Buscar el primer punto donde |sinc(x)| < limite para todo x >= ese punto
while True:
    muestra = [abs(funcion(punto + incremento * j)) for j in range(1000)]
    if all(valor < limite for valor in muestra):
        x_limite = punto
        break
    punto += incremento

# Mostrar el resultado final
print(f"El valor aproximado de x donde la condición se cumple es: {x_limite:.4f}")

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función sinc
def grafica(valores):
    return np.where(valores != 0, np.sin(valores) / valores, 1.0)

# Generar valores de entrada entre 0 y 75 con 1000 puntos
dominio = np.linspace(0, 75, 1000)
imagen = grafica(dominio)

# Configurar y mostrar el gráfico
plt.figure(figsize=(10, 5))
plt.plot(dominio, imagen, color='navy', label='sinc(x) = sin(x)/x')
plt.axhline(y=0.015, color='crimson', linestyle='--', label='Límite y = 0.015')
plt.axhline(y=0, color='gray', linestyle='--')  # Línea horizontal en y = 0
plt.axvline(x=0, color='gray', linestyle='--')  # Línea vertical en x = 0
plt.title('Representación gráfica de la función sinc modificada')
plt.xlabel('Eje x')
plt.ylabel('sinc(x)')
plt.grid(visible=True)
plt.legend()
plt.tight_layout()
plt.show()

```
