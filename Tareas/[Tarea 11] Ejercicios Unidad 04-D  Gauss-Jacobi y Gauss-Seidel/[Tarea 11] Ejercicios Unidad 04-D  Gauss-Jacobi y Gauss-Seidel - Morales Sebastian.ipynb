**1. Encuentre las primeras dos iteraciones del m√©todo de Jacobi para los siguientes sistemas lineales, por medio de ùê±^(ùüé) = 0:**

```python
import numpy as np

def jacobi_iteration(A, b, x0, max_iterations=2):
    n = len(b)
    x = x0.copy()
    iterations = [x0.copy()]
    
    for _ in range(max_iterations):
        x_new = np.zeros_like(x)
        for i in range(n):
            s = np.dot(A[i, :i], x[:i]) + np.dot(A[i, i+1:], x[i+1:])
            x_new[i] = (b[i] - s) / A[i, i]
        x = x_new
        iterations.append(x.copy())
    
    return iterations

# Definimos los sistemas lineales

# Sistema a)
A_a = np.array([[3, -1, 1],
                [3, 6, 2],
                [3, 3, 7]])
b_a = np.array([1, 0, 4])

# Sistema b)
A_b = np.array([[10, -1, 0],
                [-1, 10, -2],
                [0, -2, 10]])
b_b = np.array([9, 7, 6])

# Sistema c)
A_c = np.array([[10, 5, 0, 0],
                [5, 10, -4, 0],
                [0, -4, 8, -1],
                [0, 0, -1, 5]])
b_c = np.array([6, 25, -11, -11])

# Sistema d)
A_d = np.array([[4, 1, 1, 0, 1],
                [-1, -3, 1, 1, 0],
                [2, 1, 5, -1, -1],
                [-1, -1, -1, 4, 0],
                [0, 2, -1, 1, 4]])
b_d = np.array([6, 6, 6, 6, 6])

# Vector inicial
x0 = np.zeros(5)  # Usamos tama√±o m√°ximo para todos los sistemas

# Realizamos las iteraciones para cada sistema
print("\nResultados para los sistema con Jacobi:\n")
print("Sistema a):")
iter_a = jacobi_iteration(A_a, b_a, x0[:3])
for i, x in enumerate(iter_a):
    print(f"Iteraci√≥n {i}: {x}")

print("\nSistema b):")
iter_b = jacobi_iteration(A_b, b_b, x0[:3])
for i, x in enumerate(iter_b):
    print(f"Iteraci√≥n {i}: {x}")

print("\nSistema c):")
iter_c = jacobi_iteration(A_c, b_c, x0[:4])
for i, x in enumerate(iter_c):
    print(f"Iteraci√≥n {i}: {x}")

print("\nSistema d):")
iter_d = jacobi_iteration(A_d, b_d, x0[:5])
for i, x in enumerate(iter_d):
    print(f"Iteraci√≥n {i}: {x}")
```

**2. Repita el ejercicio 1 usando el m√©todo de Gauss-Siedel.**

```python
def gauss_seidel_iteration(A, b, x0, max_iterations=2):

    n = len(b)
    x = x0.copy()
    iterations = [x0.copy()]
    
    for _ in range(max_iterations):
        x_new = x.copy()
        for i in range(n):
            s1 = np.dot(A[i, :i], x_new[:i])
            s2 = np.dot(A[i, i+1:], x[i+1:])
            x_new[i] = (b[i] - s1 - s2) / A[i, i]
        x = x_new
        iterations.append(x.copy())
    
    return iterations

# Definimos los mismos sistemas lineales

# Sistema a)
A_a = np.array([[3, -1, 1],
                [3, 6, 2],
                [3, 3, 7]])
b_a = np.array([1, 0, 4])

# Sistema b)
A_b = np.array([[10, -1, 0],
                [-1, 10, -2],
                [0, -2, 10]])
b_b = np.array([9, 7, 6])

# Sistema c)
A_c = np.array([[10, 5, 0, 0],
                [5, 10, -4, 0],
                [0, -4, 8, -1],
                [0, 0, -1, 5]])
b_c = np.array([6, 25, -11, -11])

# Sistema d)
A_d = np.array([[4, 1, 1, 0, 1],
                [-1, -3, 1, 1, 0],
                [2, 1, 5, -1, -1],
                [-1, -1, -1, 4, 0],
                [0, 2, -1, 1, 4]])
b_d = np.array([6, 6, 6, 6, 6])

# Vector inicial
x0 = np.zeros(5)  # Usamos tama√±o m√°ximo para todos los sistemas

# Realizamos las iteraciones para cada sistema
print("\nResultados para los sistema con Gauss-Seidel:\n")
print("Sistema a):")
iter_a = gauss_seidel_iteration(A_a, b_a, x0[:3])
for i, x in enumerate(iter_a):
    print(f"Iteraci√≥n {i}: {x}")

print("\nSistema b):")
iter_b = gauss_seidel_iteration(A_b, b_b, x0[:3])
for i, x in enumerate(iter_b):
    print(f"Iteraci√≥n {i}: {x}")

print("\nSistema c):")
iter_c = gauss_seidel_iteration(A_c, b_c, x0[:4])
for i, x in enumerate(iter_c):
    print(f"Iteraci√≥n {i}: {x}")

print("\nSistema d):")
iter_d = gauss_seidel_iteration(A_d, b_d, x0[:5])
for i, x in enumerate(iter_d):
    print(f"Iteraci√≥n {i}: {x}")
```

**3. Utilice el m√©todo de Jacobi para resolver los sistemas lineales en el ejercicio 1, con TOL = 10^-3.**

```python
def jacobi(A, b, x0=None, tol=1e-3, max_iter=100):
 
    n = len(b)
    if x0 is None:
        x0 = np.zeros(n)
    x = x0.copy()
    history = [x0.copy()]
    
    for iterations in range(1, max_iter+1):
        x_new = np.zeros_like(x)
        for i in range(n):
            s = np.dot(A[i, :i], x[:i]) + np.dot(A[i, i+1:], x[i+1:])
            x_new[i] = (b[i] - s) / A[i, i]
        
        history.append(x_new.copy())
        
        # Criterio de parada: norma infinito de la diferencia
        if np.max(np.abs(x_new - x)) < tol:
            break
            
        x = x_new
    
    return x_new, iterations, history

# Definici√≥n de los sistemas lineales (los mismos del ejercicio 1)

# Sistema a)
A_a = np.array([[3, -1, 1],
                [3, 6, 2],
                [3, 3, 7]])
b_a = np.array([1, 0, 4])

# Sistema b)
A_b = np.array([[10, -1, 0],
                [-1, 10, -2],
                [0, -2, 10]])
b_b = np.array([9, 7, 6])

# Sistema c)
A_c = np.array([[10, 5, 0, 0],
                [5, 10, -4, 0],
                [0, -4, 8, -1],
                [0, 0, -1, 5]])
b_c = np.array([6, 25, -11, -11])

# Sistema d)
A_d = np.array([[4, 1, 1, 0, 1],
                [-1, -3, 1, 1, 0],
                [2, 1, 5, -1, -1],
                [-1, -1, -1, 4, 0],
                [0, 2, -1, 1, 4]])
b_d = np.array([6, 6, 6, 6, 6])

# Resoluci√≥n con tolerancia 1e-3
print("\nResultados para los sistema con Jacobi con tolerancia 1e-3:\n")
print("Sistema a):")
sol_a, iter_a, hist_a = jacobi(A_a, b_a, tol=1e-3)
print(f"Soluci√≥n: {sol_a}")
print(f"Iteraciones: {iter_a}")

print("\nSistema b):")
sol_b, iter_b, hist_b = jacobi(A_b, b_b, tol=1e-3)
print(f"Soluci√≥n: {sol_b}")
print(f"Iteraciones: {iter_b}")

print("\nSistema c):")
sol_c, iter_c, hist_c = jacobi(A_c, b_c, tol=1e-3)
print(f"Soluci√≥n: {sol_c}")
print(f"Iteraciones: {iter_c}")

print("\nSistema d):")
sol_d, iter_d, hist_d = jacobi(A_d, b_d, tol=1e-3)
print(f"Soluci√≥n: {sol_d}")
print(f"Iteraciones: {iter_d}")
```

**4. Utilice el m√©todo de Gauss-Siedel para resolver los sistemas lineales en el ejercicio 1, con TOL = 10^-3.**

```python
def gauss_seidel(A, b, x0=None, tol=1e-3, max_iter=100):

    n = len(b)
    if x0 is None:
        x0 = np.zeros(n)
    x = x0.copy()
    history = [x0.copy()]
    
    for iterations in range(1, max_iter+1):
        x_new = x.copy()
        for i in range(n):
            s1 = np.dot(A[i, :i], x_new[:i])  # Usa los valores ya actualizados
            s2 = np.dot(A[i, i+1:], x[i+1:])  # Usa los valores de la iteraci√≥n anterior
            x_new[i] = (b[i] - s1 - s2) / A[i, i]
        
        history.append(x_new.copy())
        
        # Criterio de parada: norma infinito de la diferencia
        if np.max(np.abs(x_new - x)) < tol:
            break
            
        x = x_new
    
    return x_new, iterations, history

# Definici√≥n de los sistemas lineales (los mismos del ejercicio 1)

# Sistema a)
A_a = np.array([[3, -1, 1],
                [3, 6, 2],
                [3, 3, 7]])
b_a = np.array([1, 0, 4])

# Sistema b)
A_b = np.array([[10, -1, 0],
                [-1, 10, -2],
                [0, -2, 10]])
b_b = np.array([9, 7, 6])

# Sistema c)
A_c = np.array([[10, 5, 0, 0],
                [5, 10, -4, 0],
                [0, -4, 8, -1],
                [0, 0, -1, 5]])
b_c = np.array([6, 25, -11, -11])

# Sistema d)
A_d = np.array([[4, 1, 1, 0, 1],
                [-1, -3, 1, 1, 0],
                [2, 1, 5, -1, -1],
                [-1, -1, -1, 4, 0],
                [0, 2, -1, 1, 4]])
b_d = np.array([6, 6, 6, 6, 6])

# Resoluci√≥n con tolerancia 1e-3
print("\nResultados para los sistema Gauss-Seidel con tolerancia 1e-3:\n")
print("Sistema a):")
sol_a, iter_a, hist_a = gauss_seidel(A_a, b_a, tol=1e-3)
print(f"Soluci√≥n: {sol_a}")
print(f"Iteraciones: {iter_a}")

print("\nSistema b):")
sol_b, iter_b, hist_b = gauss_seidel(A_b, b_b, tol=1e-3)
print(f"Soluci√≥n: {sol_b}")
print(f"Iteraciones: {iter_b}")

print("\nSistema c):")
sol_c, iter_c, hist_c = gauss_seidel(A_c, b_c, tol=1e-3)
print(f"Soluci√≥n: {sol_c}")
print(f"Iteraciones: {iter_c}")

print("\nSistema d):")
sol_d, iter_d, hist_d = gauss_seidel(A_d, b_d, tol=1e-3)
print(f"Soluci√≥n: {sol_d}")
print(f"Iteraciones: {iter_d}")
```
