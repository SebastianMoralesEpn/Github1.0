```python
from scipy.optimize import newton

# Definimos la función f(x)
def f(x):
    return ((x + 3) / 2)**2 - 2

# Definimos su derivada f'(x)
def f_prime(x):
    return 2 * ((x + 3) / 2) * (1/2)

# Usamos dos valores iniciales distintos para encontrar ambas soluciones
x1 = newton(f, x0=-5, fprime=f_prime)
x2 = newton(f, x0=0, fprime=f_prime)

# Mostramos los resultados
print("Las intersecciones con y = -2 ocurren en los puntos:")
print(f"({x1:.4f}, -2)")
print(f"({x2:.4f}, -2)")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import newton

# Función y derivada para Newton
def f(x):
    return ((x + 3) / 2)**2 - 2

def f_prime(x):
    return ((x + 3) / 2)

# Encontrar raíces (intersecciones con y = -2)
x1 = newton(f, x0=-5, fprime=f_prime)
x2 = newton(f, x0=0, fprime=f_prime)

# Coordenadas de los puntos de intersección
y_val = -2
points_x = [x1, x2]
points_y = [y_val, y_val]

# Valores de x para graficar la curva
x_all = np.linspace(-10, 5, 400)
# Máscara para valores válidos (donde la expresión bajo la raíz es ≥ 0)
valid_mask = ((x_all + 3)/2)**2 - 1 >= 0
x_vals = x_all[valid_mask]
# Ahora sí calculamos y sin errores
y_upper = 2 * np.sqrt(((x_vals + 3)/2)**2 - 1)
y_lower = -2 * np.sqrt(((x_vals + 3)/2)**2 - 1)

# Graficar
plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_upper, label='Curva (superior)')
plt.plot(x_vals, y_lower, label='Curva (inferior)')
plt.axhline(y=-2, color='red', linestyle='--', label='Recta y = -2')
plt.scatter(points_x, points_y, color='black', zorder=5, label='Intersecciones')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Intersección de la curva con la recta y = -2')
plt.grid(True)
plt.legend()
plt.axis('equal')
plt.show()

```
