# Algoritmo 04

```python
def algoritmo04(N):
    S_N = 0.0
    for n in range(1, N + 1):
        S_N += 1 / n
    return S_N

# Ejemplo de uso
N = 10000  # N puede variar para calcular la serie armónica hasta otro N
resultado = algoritmo04(N)
print(f"La suma de la serie geométrica hasta N={N} es: {resultado}")
```

**Respuesta:** A medida que (N) aumenta, la suma tenderá al infinito.
