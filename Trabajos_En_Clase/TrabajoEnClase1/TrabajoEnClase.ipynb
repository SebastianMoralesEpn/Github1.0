# **Graficar la siguiente ecuación:**



```python
import matplotlib.pyplot as plt
import numpy as np

def equation(x: float) -> float:
    return 1/4 * (x**3 + 3*x**2 - 6*x - 8)
    
# Generate x values with more points
x = np.linspace(-10, 10, 100)  # Cambiado de 10 a 100 puntos

# Calculate y values
y = equation(x)

# Plot the equation
plt.plot(x, y, linestyle='dotted')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of $\\frac{1}{4}(x^3 + 3x^2 - 6x - 8) = 0$')
ax = plt.gca()
ax.set_ylim([-10, 10])
ax.set_xlim([-5, 3])
plt.grid(True)
plt.show()
```

```python
def bisection_method(func, a, b, tol=1e-5, max_iter=100):
    """
    Método de bisección para encontrar una raíz de la función `func` en el intervalo [a, b].
    
    Parameters:
        func (callable): La función para la cual se busca la raíz.
        a (float): Extremo izquierdo del intervalo.
        b (float): Extremo derecho del intervalo.
        tol (float): Tolerancia para el criterio de convergencia.
        max_iter (int): Número máximo de iteraciones.
    
    Returns:
        tuple: La raíz aproximada de la función y el número de iteraciones.
    """
    if func(a) * func(b) >= 0:
        raise ValueError("La función debe cambiar de signo en el intervalo [a, b].")

    iter_count = 0
    while (b - a) / 2 > tol and iter_count < max_iter:
        c = (a + b) / 2
        if func(c) == 0:  # Encontramos la raíz exacta
            return c, iter_count
        elif func(a) * func(c) < 0:
            b = c
        else:
            a = c
        iter_count += 1

    return (a + b) / 2, iter_count  # Retorna la raíz aproximada y el número de iteraciones

# Usar el método de bisección para encontrar una raíz de la ecuación
a, b = -4.7, 2.5  # Intervalo inicial
root, iter_count = bisection_method(equation, a, b)

print(f"La raíz aproximada encontrada es: {root}")
print(f"El número de iteraciones necesarias fue: {iter_count}")
```
