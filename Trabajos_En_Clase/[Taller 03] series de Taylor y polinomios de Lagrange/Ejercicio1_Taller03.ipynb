##### **Grafique las curvas de las series de Taylor de varios órdenes para los siguientes casos:**

##### 1. $cos(x)$, $x_0 = 0$

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-np.pi, np.pi, 100)
cos_aprox = 1
cos_aprox1 = 1 - (1/2) * x**2
cos_aprox2 = 1 - (1/2) * x**2 + (1/24) * x**4

plt.figure(figsize=(10, 6))
plt.plot(x, np.cos(x), label='cos(x)', color='blue')
plt.plot(x, [cos_aprox]*len(x), label='cos(x) = 1', linestyle='--', color='red')
plt.plot(x, cos_aprox1, label='cos(x) = 1 - (1/2)x^2', linestyle='--', color='green')
plt.plot(x, cos_aprox2, label='cos(x) = 1 - (1/2)x^2 + (1/24)x^4', linestyle='--', color='orange')

plt.title('Taylor Series Approximation of cos(x)')
plt.xlabel('x')
plt.ylabel('cos(x)')
plt.legend()
plt.grid(True)
plt.xlim(-np.pi, np.pi)
plt.ylim(-1.5, 1.5)
plt.show()
```
