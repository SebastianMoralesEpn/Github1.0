```python
# Derivadas parciales para regresión lineal
# ####################################################################
def der_parcial_1(xs: list, ys: list) -> tuple[float, float, float]:
    """Retorna los coeficientes de la ecuación de la derivada parcial con respecto al parámetro 1 al reemplazar los valores ``xs`` y ``ys``. La ecuación es de la forma:
    c_1 * a_1 + c_0 * a_0 = c_ind


    ## Parameters

    ``xs``: lista de valores de x.

    ``ys``: lista de valores de y.

    ## Return

    ``c_1``: coeficiente del parámetro 1.

    ``c_0``: coeficiente del parámetro 0.

    ``c_ind``: coeficiente del término independiente.

    """

    # coeficiente del término independiente
    c_ind = sum(ys)

    # coeficiente del parámetro 1
    c_1 = sum(xs)

    # coeficiente del parámetro 0
    c_0 = len(xs)

    return (c_1, c_0, c_ind)


def der_parcial_0(xs: list, ys: list) -> tuple[float, float, float]:
    """Retorna los coeficientes de la ecuación de la derivada parcial con respecto al parámetro 0 al reemplazar los valores ``xs`` y ``ys``. La ecuación es de la forma:
    c_1 * a_1 + c_0 * a_0 = c_ind


    ## Parameters

    ``xs``: lista de valores de x.

    ``ys``: lista de valores de y.

    ## Return

    ``c_1``: coeficiente del parámetro 1.

    ``c_0``: coeficiente del parámetro 0.

    ``c_ind``: coeficiente del término independiente.

    """
    c_1 = 0
    c_0 = 0
    c_ind = 0
    for xi, yi in zip(xs, ys):
        # coeficiente del término independiente
        c_ind += xi * yi

        # coeficiente del parámetro 1
        c_1 += xi * xi

        # coeficiente del parámetro 0
        c_0 += xi

    return (c_1, c_0, c_ind)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Construir gradientes para polinomio de grado 4
def der_parcial_k(xs, ys, k):
    """Derivada parcial respecto al parámetro k para ajuste polinómico de grado 4"""
    n = 5
    c = []
    for i in range(n):
        c_i = sum(x**(i+k) for x in xs)
        c.append(c_i)
    c_ind = sum(y * (x**k) for x, y in zip(xs, ys))
    return (*c, c_ind)

gradiente = [
    lambda xs, ys: der_parcial_k(xs, ys, 0),
    lambda xs, ys: der_parcial_k(xs, ys, 1),
    lambda xs, ys: der_parcial_k(xs, ys, 2),
    lambda xs, ys: der_parcial_k(xs, ys, 3),
    lambda xs, ys: der_parcial_k(xs, ys, 4),
]

from src import ajustar_min_cuadrados

# Ajuste por mínimos cuadrados usando las derivadas parciales para grado 4
parametros = ajustar_min_cuadrados(xs, ys, gradiente=gradiente)

# Crear valores para graficar el polinomio ajustado
x_linea = np.linspace(min(xs), max(xs), 100)
y_linea = sum(parametros[i] * x_linea**i for i in range(5))

def modelo(x):
    return sum(parametros[i] * x**i for i in range(5))

# Graficar los datos y el polinomio ajustado
plt.scatter(xs, ys, label='Datos')
plt.plot(x_linea, y_linea, color='red', label='Polinomio ajustado')
plt.xlabel('Valores de X')
plt.ylabel('Valores de Y')
plt.title('Ajuste polinómico de grado 4 por mínimos cuadrados')
plt.legend()
plt.show()

# Calcular el error relativo promedio y el máximo error relativo
errores_relativos = [abs(ys[j] - modelo(xs[j])) / abs(ys[j]) for j in range(len(xs))]
error_promedio = np.mean(errores_relativos)
error_maximo = np.max(errores_relativos)

# Imprimir el polinomio ajustado y los errores
print(f'Polinomio ajustado: y = {parametros[4]:.4f} x^4 + {parametros[3]:.4f} x^3 + {parametros[2]:.4f} x^2 + {parametros[1]:.4f} x + {parametros[0]:.4f}')
print(f'Error relativo promedio: {error_promedio}')
print(f'Máximo error relativo: {error_maximo}')
```
